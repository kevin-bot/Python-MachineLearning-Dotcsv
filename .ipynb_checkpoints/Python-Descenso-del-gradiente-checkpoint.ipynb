# GRADIENT DESCENT

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
fun = lambda th : np.sin(1/2*th[0]**2 - 1/4*th[1]**2 + 3) * np.cos(2*th[0]+1-np.e**th[1])

res = 100

_x =np.linspace(-2,2, res)
_y =np.linspace(-2,2, res)

_z = np.zeros((res,res))

for ix, x in enumerate(_x):
    for iy, y in enumerate (_y):
        _z[iy,ix] = fun([x,y])
        
```

```python
plt.contourf(_x,_y,_z,100)
plt.colorbar()


#el punto
theta = np.random.rand(2) * 4 -2

#derivadas parciales
_t = np.copy(theta)

h=0.001
radio_aprem = 0.001

plt.plot(theta[0],theta[1], "o", c= "white")

grad = np.zeros(2)

for repite in range(10000):
    for it, th in enumerate (theta):
        _t = np.copy(theta)
        
        _t[it] =  _t[it] + h

        deriv = (fun(_t) - fun(theta)) / h
        grad [it] = deriv

    theta = theta - radio_aprem*grad
    
    if(repite%100 == 0):
        plt.plot(theta[0],theta[1], ".", c= "red")
    
plt.plot(theta[0],theta[1], "o", c= "green")
plt.show()
```
